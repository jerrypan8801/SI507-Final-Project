```python
game_data = game
champ_network = ChampionNetwork(game_data=game_data)

while True:
    print("\n1. Query a Champion")
    print("2. View Most Common Champion Pairs")
    print("3. Most Influential Champions")
    print("4. Find Isolated Champions")

    choice = input("Choose an option or 'q' to quit: ")

    if choice == "q":
        break
    elif choice == "1":
        champ_id = int(input("Enter Champion ID: "))
        print(champ_network.query_champion(champ_id))
    elif choice == "2":
        print(champ_network.common_pairs())
    elif choice == "3":
        print(champ_network.most_influential_champions())
    elif choice == '4':
        isolated = champ_network.find_isolated_champions()
        if not isolated:
            print("No isolated champions found.")
        else:
            print("Isolated Champions:", isolated)
```
